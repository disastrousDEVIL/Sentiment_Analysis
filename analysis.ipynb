Stopwords are commonly occurring words (e.g., the, and, is, in) that typically don’t add meaningful information in text analysis.

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
import string
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    ps = PorterStemmer()

    y = []
    for i in text:
        if i.isalnum():
            y.append(i)

    text = y[:]
    y.clear()

    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)

    text = y[:]
    y.clear()

    for i in text:
        y.append(ps.stem(i))


    return " ".join(y)
```

 What is Stemming?
Stemming is the process of reducing words to their root or stem form.
It removes common endings or prefixes (suffixes/prefixes).
How does PorterStemmer work?
Applies predefined rules to trim suffixes.
Reduces different forms of a word to a common stem (not always a complete dictionary word).

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python

gnb.fit(X_train,y_train)
y_pred1 = gnb.predict(X_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```
